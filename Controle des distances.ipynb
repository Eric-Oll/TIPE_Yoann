LinearRoad

```python
%pylab inline
from roadmaps.linear_road import LinearRoad
from parameters import DISTANCE

D = lambda x1, x2: x2-x1
NB_POINTS = 1000
nb_roadmap = 1

fig, ax = plt.subplots(nb_roadmap+1, 1, figsize=(25,5*(nb_roadmap+1)+2), squeeze=False, gridspec_kw={'hspace': 0.5})
fig.set_facecolor('w')

linear_road = LinearRoad(ax[0,0])
#ax[0].set_visible(False)
linear_road.landscape()
ax[0,0].plot([pos.x for pos in linear_road.roadmap[0]], [pos.y for pos in linear_road.roadmap[0]], 'g.-')
X = np.linspace(*ax[0,0].get_xlim(), NB_POINTS)
ax[0,0].plot(X, [-0.5]*NB_POINTS, 'r.')

for num_path, roadmap in enumerate(linear_road.roadmap):
    ax[num_path+1,0].margins(0,0)
    ax[num_path+1,0].plot(range(1, len(roadmap)-1), [DISTANCE(roadmap[i], roadmap[i+1]) for i in range(len(roadmap)-2)])
ax[1,0].plot(range(1, len(X)-1),[D(X[i], X[i+1]) for i in range(len(X)-2)], 'r--')
    
plt.show()
```
